# Churn Rate Overview

```python
# Churn Rate
churn_rate=merge_df["Customer Status"].value_counts(normalize=True)*100
print(churn_rate)
```

```python
# Churn Rate by City
churn_rate_city=merge_df[merge_df["Customer Status"]=="Churned"].groupby("City").size().sort_values(ascending=False)
print(churn_rate_city.head(10))
```

```python
# Churn Rate by Internet Type
churn_rate_it=merge_df.groupby("Internet Type")["Customer Status"].value_counts(normalize=True).unstack().fillna(0)*100
print(churn_rate_it)
```

```python
# Churn Rate by Age Group
def get_age_group(age):
    if age<30:
        return '<30'
    elif 30<= age<50:
        return '30-50'
    elif 50<= age< 70:
        return '50-70'
    else:
        return '70+'
merge_df["Age Group"] = merge_df["Age"].apply(get_age_group)
churn_by_age=merge_df[merge_df["Customer Status"]=="Churned"]["Age Group"].value_counts().sort_index()
print(churn_by_age)
```

```python
# Churn Rate by Contract Type
churn_by_contract=merge_df.groupby("Contract")["Customer Status"].value_counts(normalize=True).unstack().fillna(0)*100
print(churn_by_contract)
```

```python
# Correlation with Revenue
merge_df.groupby("Customer Status")["Total Revenue"].mean()
```

# Visualisation

```python
plt.figure(figsize=(6, 5))
sns.barplot(x=churn_by_age.index, y=churn_by_age.values, palette="Set2")
plt.title('Churned Customers by Age Group', fontsize=14)
plt.xlabel('Age Group')
plt.ylabel('Number of Churned Customers')
plt.grid(axis='y', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Churn Rate by Internet Type [Visualisation]

```python

churn_by_internet = merge_df.groupby('Internet Type').apply(
    lambda x: (x['Customer Status'] == 'Churned').sum() * 100 / len(x)
).reset_index(name='Churn Rate')
```

```python

plt.figure(figsize=(6, 5))
sns.barplot(data=churn_by_internet, x='Internet Type', y='Churn Rate', palette='pastel')
plt.title('Churn Rate by Internet Type', fontsize=14)
plt.xlabel('Internet Type')
plt.ylabel('Churn Rate (%)')
plt.xticks(rotation=20)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Customer Status Distribution

```python
status_counts = merge_df['Customer Status'].value_counts()
```

```python
plt.figure(figsize=(5, 5))
colors = ['#FF9999','#66B3FF','#99FF99']
status_counts.plot.pie(
    autopct='%1.1f%%', 
    startangle=140, 
    colors=colors, 
    shadow=True,
    explode=(0.05, 0.05, 0.05)  
)
plt.title('Customer Status Distribution', fontsize=14)
plt.ylabel('')
plt.tight_layout()
plt.show()
```
